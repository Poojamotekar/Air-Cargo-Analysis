```python
def Replace_values(x):
      if x=='Alone':
         return 'Single'
      elif x=='Absurd':
         return 'Single'
      elif x=='YOLO':
        return 'Together'
      else:
       return x

df_clean['Marital_Status']=df_clean['Marital_Status'].apply(Replace_values)
```
